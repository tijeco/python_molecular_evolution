## Exponential distribution

The probability distribution of the time between events in a Poisson point process.

$$f(x) = \lambda e^{-\lambda x}$$

$$mean = \frac{1}{\lambda}$$
$$variance = \frac{1}{\lambda^2}$$




````python
%matplotlib inline


def poisson_n(λ,n):
    x = np.arange(0, n, 1)
    y = np.exp(-λ)*np.power(λ, x)/factorial(x)
    return x,y


import numpy as np
import matplotlib.pyplot as plt
def exponential_n(λ,n):
    x = np.arange(0, n,0.1)  
    y = (λ) * np.exp(-x * λ)
    return x, y

        
exponential_05_x, exponential_05_y = exponential_n(0.5,5)
exponential_10_x, exponential_10_y = exponential_n(1,5)
exponential_15_x, exponential_15_y = exponential_n(1.5,5)

plt.plot(exponential_05_x, exponential_05_y,"b",label="λ=0.5")
plt.plot(exponential_10_x, exponential_10_y,"g",label="λ=1.0")
plt.plot(exponential_15_x, exponential_15_y,"k",label="λ=1.5")

plt.yticks([0,0.5,1.0,1.5])
plt.legend()
plt.show()
````


## Gamma distribution

$$f(x;\alpha,\beta) = \frac{\beta^{\alpha}}{\Gamma(\alpha)}e^{-\beta  x} x^{\alpha -1}, \alpha > 0,\beta > 0, x > 0$$

$$ \Gamma (\alpha)=(\alpha-1)!$$

$$ mean = \frac{\alpha}{\beta}$$
$$ variance = \frac{\alpha}{\beta^2}$$

Where $\alpha$ is the shape and $\beta$ is the scale


````python
%matplotlib inline
from scipy.special import factorial
def Γ(x):
    return factorial(x-1)
def gamma_n(β,α,n):
    x = np.arange(0, n, 0.1)
    y = ((β**α)/Γ(α)) * np.exp(-β*x) * x**(α-1)
    return x,y

gamma_b1_a3_x, gamma_b1_a3_y = gamma_n(1,3,20)
gamma_b2_a2_x, gamma_b2_a2_y = gamma_n(2,2,20)
gamma_b3_a1_x, gamma_b3_a1_y = gamma_n(3,1,20)

plt.plot(gamma_b1_a3_x, gamma_b1_a3_y)
plt.plot(gamma_b2_a2_x, gamma_b2_a2_y)
plt.plot(gamma_b3_a1_x, gamma_b3_a1_y)
plt.show()
````

## Beta distribution

$$ f(x; \alpha,\beta) = \frac{1}{B(\alpha,\beta)}  x^{\alpha-1}(1-x)^{\beta-1}, 0 <x<1,\alpha>0,\beta>0  $$


$$ B(\alpha,\beta)  = \frac{\Gamma(\alpha)\Gamma(\beta)}{\Gamma(\alpha + \beta)}$$

$$ mean =  \frac{\alpha}{\alpha + \beta}$$
$$ variance = \frac{\alpha \beta}{(\alpha + \beta)^2(\alpha + \beta + 1)} $$

````python
def B(α,β):
    return (Γ(α)*Γ(β))/(Γ(α+β))
def beta_distribution(α,β): # to be changed to alpha and beta
    x = np.arange(0, 1, 0.05)
    y = 1/(B(α,β)) * x**(α -1) * (1 - x)**(β -1)
    return x,y


# plt.plot(x,y)
# plt.show
````
