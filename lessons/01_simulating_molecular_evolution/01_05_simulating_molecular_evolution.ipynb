## Convert phylogenetic tree to python dictionary

The following tree has 4 tips (a,b,c,d) and four nodes (n0,n2,n3,n4). The time scale indicates how many years (in millions of years) divergences occured before present time.



Construct a dictionary with keywords ```children```,```parent```, and ```time``` for each node.

````python
phylogeny = {
    "children":{
        "n1":{
            "children":{
                "a":{
                    "children":None,
                    "time":6,
                    "parent":"n1"
                },
                "n2":{
                    "children":{
                        "b":{
                            "children":None,
                            "time":4,
                            "parent":"n2"
                        },
                        "n3":{
                            "children":{
                                "c":{
                                    "children":None,
                                    "time":2,
                                    "parent":"n3"
                                },
                                "d":{
                                    "children":None,
                                    "time":2,
                                    "parent":"n3"
                                }
                            },
                            "time":2,
                            "parent":"n2"
                        }
                    },
                    "time":2,
                    "parent":"n1"
                }
            },
            "time":None,
            "parent":None
            
        }
    },
    "time":None,
    "parent":None
  
}
````

### Guided excersice

Iterate over phylogeny recursively to print how long each node diverged from its ancestor.

````python
def get_children(phy):
    parent = None
    if phy["children"] == None:
        return phy["time"],phy["parent"]
    else:
        for child in phy["children"]:
            if get_children(phy["children"][child]) == None: 
                time = phy["children"][child]["time"]
                parent = phy["children"][child]["parent"]
            else:
                
                time,parent = get_children(phy["children"][child])
            print(child,"evolved for",time,"million years from",parent)
            
print(get_children(phylogeny))
            

    
        
````
